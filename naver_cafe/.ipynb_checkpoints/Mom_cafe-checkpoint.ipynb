```python
real_url = "https://m.cafe.naver.com/ArticleSearchListAjax.nhn?search.query=%EB%B6%84%EC%9C%A0&search.menuid=&search.searchBy=1&search.sortBy=date&search.clubid=10094499&search.option=0&search.defaultValue=&search.page="
```

```python
def cafe_crawler(url) :
    total_text = []
    for i in range(50) :
        now_url = url + str(i+1)
        temp = requests.get(now_url)
        contents = BeautifulSoup(temp.content, "html.parser")
        tag = contents.select('.list_tit li div h3')
        #print(now_url)
        for line in tag :
            total_text.append(line.text)
        
    return(total_text)
```

```python
titles = cafe_crawler(real_url)
```
